```python
%matplotlib inline
import numpy as np
from numpy import zeros, ones, shape, dot, multiply
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
citi_names=[]
with open('cities.txt') as f:
    for line in f:
        citi_names.append(line.strip())
citi_names = citi_names[:10]
```

```python
D=[]
with open('distances.txt') as f:
    for line in f:
        D.append(map(float,line.strip().split(',')))
D = np.array(D)
D = D*D
```

```python
H = np.eye(10)-1/10.*np.outer(np.ones(10),np.ones(10))
```

```python
B = -0.5*dot(dot(H,D),H)
```

```python
U,s,V=np.linalg.svd(B,full_matrices=False)
```

```python
data=dot(U,np.diag(np.sqrt(s)))[:,:2]
data=-data
```

```python
plt.title('Classical Multidimensional Scaling')
plt.scatter(data[:,0],data[:,1])
for i, txt in enumerate(citi_names):
    plt.annotate(txt, (data[:,0][i],data[:,1][i]),fontsize=15)
plt.show()
```

```python
s,U=np.linalg.eig(B)
```

```python
data=dot(U,np.diag(np.sqrt(s)))[:,:2]
```

```python
plt.title('Classical Multidimensional Scaling')
plt.scatter(data[:,0],data[:,1])
for i, txt in enumerate(citi_names):
    plt.annotate(txt, (data[:,0][i],data[:,1][i]),fontsize=15)
plt.show()
```
